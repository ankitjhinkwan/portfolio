## Step 1 — Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Cyberpunk dark style
plt.style.use('dark_background')
plt.rcParams.update({
    'figure.facecolor': '#050510',
    'axes.facecolor':   '#0a0a1a',
    'axes.edgecolor':   '#00fff9',
    'axes.labelcolor':  '#00fff9',
    'xtick.color':      '#00fff9',
    'ytick.color':      '#00fff9',
    'text.color':       '#00fff9',
    'grid.color':       'rgba(0,255,249,0.1)',
    'figure.figsize':   (14, 5)
})
CYAN  = '#00fff9'
PINK  = '#ff00c8'
YELLO = '#ffe600'
GREEN = '#00ff41'
print('⚡ Cyberpunk mode activated!')
```

## Step 2 — Load & Explore

```python
df = pd.read_csv('../data/netflix_data.csv')
print(f'Total Titles:  {len(df):,}')
print(f'Movies:        {df[df["Type"]=="Movie"].shape[0]:,}')
print(f'TV Shows:      {df[df["Type"]=="TV Show"].shape[0]:,}')
print(f'Genres:        {df["Genre"].nunique()}')
print(f'Countries:     {df["Country"].nunique()}')
print(f'Year Range:    {df["ReleaseYear"].min()} — {df["ReleaseYear"].max()}')
print(f'Avg IMDB:      {df["IMDBScore"].mean():.2f}')
df.head()
```

```python
print('Missing values:')
print(df.isnull().sum())
df.describe().round(2)
```

## Step 3 — Content Type Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.patch.set_facecolor('#050510')

# Pie chart
ct = df['Type'].value_counts()
axes[0].pie(ct.values, labels=ct.index, colors=[CYAN, PINK],
            autopct='%1.1f%%', startangle=90,
            wedgeprops=dict(edgecolor='#050510', linewidth=3),
            textprops=dict(color='white', fontsize=11))
axes[0].set_title('Content Type Split', color=CYAN, fontsize=12, fontweight='bold')
axes[0].set_facecolor('#050510')

# IMDB by type
for t, color in [('Movie', CYAN), ('TV Show', PINK)]:
    axes[1].hist(df[df['Type']==t]['IMDBScore'], bins=25, alpha=0.7, label=t, color=color, edgecolor='#050510')
axes[1].set_title('IMDB Score by Type', color=CYAN, fontsize=12, fontweight='bold')
axes[1].set_xlabel('IMDB Score'); axes[1].legend()
axes[1].set_facecolor('#0a0a1a')

# Rating distribution
ratings = df['Rating'].value_counts().head(8)
axes[2].barh(ratings.index, ratings.values, color=CYAN, edgecolor='#050510')
axes[2].set_title('Content Ratings', color=CYAN, fontsize=12, fontweight='bold')
axes[2].set_facecolor('#0a0a1a')
for i, v in enumerate(ratings.values):
    axes[2].text(v+5, i, str(v), va='center', color=PINK, fontweight='bold')

plt.tight_layout(); plt.savefig('../content_type.png', dpi=150, bbox_inches='tight'); plt.show()
```

## Step 4 — Genre Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.patch.set_facecolor('#050510')

# Genre counts
genre_counts = df['Genre'].value_counts()
colors = [CYAN if i%2==0 else PINK for i in range(len(genre_counts))]
axes[0].barh(genre_counts.index[::-1], genre_counts.values[::-1], color=colors[::-1], edgecolor='#050510')
axes[0].set_title('Titles by Genre', color=CYAN, fontsize=12, fontweight='bold')
axes[0].set_facecolor('#0a0a1a')
for i, v in enumerate(genre_counts.values[::-1]):
    axes[0].text(v+5, i, str(v), va='center', color=YELLO, fontweight='bold', fontsize=9)

# Genre avg IMDB
genre_imdb = df.groupby('Genre')['IMDBScore'].mean().sort_values(ascending=True)
bar_colors = plt.cm.cool(np.linspace(0, 1, len(genre_imdb)))
axes[1].barh(genre_imdb.index, genre_imdb.values, color=bar_colors, edgecolor='#050510')
axes[1].set_title('Avg IMDB Score by Genre', color=CYAN, fontsize=12, fontweight='bold')
axes[1].set_facecolor('#0a0a1a')
axes[1].axvline(df['IMDBScore'].mean(), color=PINK, linestyle='--', lw=2, label=f'Overall Avg: {df["IMDBScore"].mean():.1f}')
axes[1].legend()
for i, v in enumerate(genre_imdb.values):
    axes[1].text(v+0.05, i, f'{v:.1f}', va='center', color=YELLO, fontsize=9)

plt.tight_layout(); plt.savefig('../genre_analysis.png', dpi=150, bbox_inches='tight'); plt.show()
```

## Step 5 — Country Analysis

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.patch.set_facecolor('#050510')

country_counts = df['Country'].value_counts()
bar_c = plt.cm.plasma(np.linspace(0.2, 0.9, len(country_counts)))
axes[0].bar(country_counts.index, country_counts.values, color=bar_c, edgecolor='#050510')
axes[0].set_title('Content by Country', color=CYAN, fontsize=12, fontweight='bold')
axes[0].set_facecolor('#0a0a1a'); axes[0].tick_params(axis='x', rotation=45)
for i,v in enumerate(country_counts.values):
    axes[0].text(i, v+5, str(v), ha='center', color=CYAN, fontweight='bold', fontsize=8)

country_imdb = df.groupby('Country')['IMDBScore'].mean().sort_values(ascending=False)
axes[1].bar(country_imdb.index, country_imdb.values, color=plt.cm.cool(np.linspace(0,1,len(country_imdb))), edgecolor='#050510')
axes[1].set_title('Avg IMDB by Country', color=CYAN, fontsize=12, fontweight='bold')
axes[1].set_facecolor('#0a0a1a'); axes[1].tick_params(axis='x', rotation=45)
axes[1].axhline(df['IMDBScore'].mean(), color=PINK, linestyle='--', lw=2)
plt.tight_layout(); plt.savefig('../country_analysis.png', dpi=150, bbox_inches='tight'); plt.show()
```

## Step 6 — IMDB Score Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.patch.set_facecolor('#050510')

# IMDB histogram
axes[0].hist(df['IMDBScore'], bins=30, color=CYAN, edgecolor='#050510', alpha=0.8)
axes[0].axvline(df['IMDBScore'].mean(), color=PINK, lw=2.5, linestyle='--', label=f'Mean: {df["IMDBScore"].mean():.1f}')
axes[0].axvline(df['IMDBScore'].median(), color=YELLO, lw=2, linestyle=':', label=f'Median: {df["IMDBScore"].median():.1f}')
axes[0].set_title('IMDB Distribution', color=CYAN, fontsize=12, fontweight='bold')
axes[0].set_facecolor('#0a0a1a'); axes[0].legend()

# Top rated genres
top_rated = df.groupby('Genre')['IMDBScore'].mean().nlargest(10).sort_values()
axes[1].barh(top_rated.index, top_rated.values, color=plt.cm.cool(np.linspace(0.3,1,10)), edgecolor='#050510')
axes[1].set_title('Top Rated Genres', color=CYAN, fontsize=12, fontweight='bold')
axes[1].set_facecolor('#0a0a1a')
for i,v in enumerate(top_rated.values): axes[1].text(v+0.05, i, f'{v:.2f}', va='center', color=YELLO, fontsize=9)

# Score vs Year
year_imdb = df.groupby('ReleaseYear')['IMDBScore'].mean()
axes[2].plot(year_imdb.index, year_imdb.values, color=CYAN, lw=2.5, marker='o', ms=5)
axes[2].fill_between(year_imdb.index, year_imdb.values, alpha=0.2, color=CYAN)
axes[2].set_title('IMDB Score Over Years', color=CYAN, fontsize=12, fontweight='bold')
axes[2].set_facecolor('#0a0a1a')

plt.tight_layout(); plt.savefig('../imdb_analysis.png', dpi=150, bbox_inches='tight'); plt.show()
```

## Step 7 — Year Trends

```python
yearly = df.groupby(['ReleaseYear','Type']).size().unstack(fill_value=0)
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.patch.set_facecolor('#050510')

if 'Movie' in yearly.columns:
    axes[0].fill_between(yearly.index, yearly.get('Movie',0), alpha=0.6, color=CYAN, label='Movies')
    axes[0].plot(yearly.index, yearly.get('Movie',0), color=CYAN, lw=2)
if 'TV Show' in yearly.columns:
    axes[0].fill_between(yearly.index, yearly.get('TV Show',0), alpha=0.6, color=PINK, label='TV Shows')
    axes[0].plot(yearly.index, yearly.get('TV Show',0), color=PINK, lw=2)
axes[0].set_title('Content Added Per Year', color=CYAN, fontsize=12, fontweight='bold')
axes[0].set_facecolor('#0a0a1a'); axes[0].legend()

# Genre trend over years
top5 = df['Genre'].value_counts().nlargest(5).index
cols5 = [CYAN, PINK, YELLO, GREEN, '#a855f7']
for genre, color in zip(top5, cols5):
    gy = df[df['Genre']==genre].groupby('ReleaseYear').size()
    axes[1].plot(gy.index, gy.values, marker='o', label=genre, color=color, lw=2, ms=4)
axes[1].set_title('Top 5 Genre Trends', color=CYAN, fontsize=12, fontweight='bold')
axes[1].set_facecolor('#0a0a1a'); axes[1].legend(fontsize=8)
plt.tight_layout(); plt.savefig('../year_trends.png', dpi=150, bbox_inches='tight'); plt.show()
```
